# Noise Filtering on Corrupted Image

- (a) Apply Gaussian smoothing
- (b) Apply median filtering

```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Load the Noisy Image

```python
# Load emma.jpg
if img is None:
    img = cv2.imread('img/emma.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("emma.jpg not found")

print(f"Image size: {img.shape}")
print(f"Data type: {img.dtype}")

# Display original noisy image
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image (with Salt & Pepper Noise)', fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## (a) Gaussian Smoothing

```python
# Apply Gaussian smoothing
# Using kernel size 5x5 and sigma=1.0
img_gaussian = cv2.GaussianBlur(img, (5, 5), 1.0)

print(f"Gaussian filtered image size: {img_gaussian.shape}")

# Display result
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Original (Noisy)', fontsize=13, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_gaussian, cmap='gray', vmin=0, vmax=255)
axes[1].set_title('Gaussian Smoothing (5×5, σ=1.0)', fontsize=13, fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

## (b) Median Filtering

```python
# Apply median filtering
# Using kernel size 5x5
img_median = cv2.medianBlur(img, 5)

print(f"Median filtered image size: {img_median.shape}")

# Display result
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Original (Noisy)', fontsize=13, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_median, cmap='gray', vmin=0, vmax=255)
axes[1].set_title('Median Filtering (5×5)', fontsize=13, fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

## Display All Results

```python
# Display all three images together
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Original (Noisy)', fontsize=13, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_gaussian, cmap='gray', vmin=0, vmax=255)
axes[1].set_title('Gaussian Smoothing', fontsize=13, fontweight='bold')
axes[1].axis('off')

axes[2].imshow(img_median, cmap='gray', vmin=0, vmax=255)
axes[2].set_title('Median Filtering', fontsize=13, fontweight='bold')
axes[2].axis('off')

plt.suptitle('Noise Reduction Results', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```
